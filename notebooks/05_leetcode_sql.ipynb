**Solution + Explanation:**

**Brute force O(n²):** two nested loops, check every pair. Too slow for large inputs.

**Optimal O(n) — hash map:**
As you scan each number, store it in a dict: `{value: index}`.
For each new number, check if `target - num` is already in the dict.
If yes, you found the pair. If no, add the current number to the dict.

```python
def two_sum(nums, target):
    seen = {}                          # value -> index
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
```

**Why this works:** you only need one pass. By the time you reach the second number of a pair,
the first is already in `seen`. Time: O(n). Space: O(n).

**Interview tip:** always mention the O(n²) brute force first, then explain why O(n) is better.
Shows you think about complexity, not just correctness.

````python
# Run the optimal solution
def two_sum(nums, target):
    seen = {}
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i

print(two_sum([2, 7, 11, 15], 9))   # [0, 1]
print(two_sum([3, 2, 4], 6))        # [1, 2]
print(two_sum([3, 3], 6))           # [0, 1]
````

def is_anagram(s, t):

**Solution + Explanation:**

**Key insight:** track the minimum price seen so far.
At each day, the best profit *if you sell today* = `price[today] - min_so_far`.
Update the global max profit as you go.

```python
def max_profit(prices):
    min_price  = float('inf')
    max_profit = 0
    for price in prices:
        min_price  = min(min_price, price)
        max_profit = max(max_profit, price - min_price)
    return max_profit
```

Time: O(n). Space: O(1). One pass.

**Why this pattern matters for data science:**
This is a classic "running minimum + comparison" pattern. You'll use this same
structure when scanning time-series for: first threshold crossing, best entry point,
cumulative maximum/minimum. Recognising it as a generalised pattern is more valuable
than memorising this specific problem.

````python
def max_profit(prices):
    min_price  = float('inf')
    max_profit = 0
    for price in prices:
        min_price  = min(min_price, price)
        max_profit = max(max_profit, price - min_price)
    return max_profit

print(max_profit([7, 1, 5, 3, 6, 4]))   # 5
print(max_profit([7, 6, 4, 3, 1]))       # 0
print(max_profit([2, 4, 1]))             # 2
````

## SQL Problem — Duplicate Job Listings (DataLemur Easy)

**Source:** DataLemur — `linkedin_job_postings` table

**Problem:**
Assume you are given the table below on job listings from LinkedIn.
Write a query to find the number of companies that have posted **duplicate job listings**.

Definition: duplicate = same company_id AND same title AND same description.

```
job_listings table:
| job_id | company_id | title          | description |
|--------|-----------|----------------|-------------|
| 248    | 827        | Business Analyst | ...        |
| 149    | 845        | Business Analyst | ...        |
| 945    | 345        | Data Analyst   | ...         |
| 164    | 345        | Data Analyst   | ...         |
| 172    | 244        | Data Engineer  | ...         |
```

Expected output:
| duplicate_companies |
|--------------------|
| 1                  |   (only company 345 has a duplicate)
```

**Your solution (write SQL as a string, explain your logic):**

````python
# Write your SQL logic as a comment/explanation, then the query as a string

# Step 1: group by (company_id, title, description) and count rows per group
# Step 2: keep only groups with count > 1 (those are duplicates)
# Step 3: count distinct company_ids from step 2

solution_sql = """
SELECT COUNT(DISTINCT company_id) AS duplicate_companies
FROM (
    SELECT company_id
    FROM   job_listings
    GROUP BY company_id, title, description
    HAVING COUNT(*) > 1
) AS dupes;
"""

print(solution_sql)
````

**Solution Walkthrough:**

```sql
SELECT COUNT(DISTINCT company_id) AS duplicate_companies
FROM (
    SELECT   company_id
    FROM     job_listings
    GROUP BY company_id, title, description
    HAVING   COUNT(*) > 1
) AS dupes;
```

**Step by step:**
1. **Inner query** — `GROUP BY company_id, title, description`:
   groups rows that have the same company + title + description together.
2. **`HAVING COUNT(*) > 1`** — filters to groups with more than 1 row = actual duplicates.
3. **Outer query** — `COUNT(DISTINCT company_id)` counts how many *unique* companies
   appear in the duplicate list. (A company could have multiple pairs of duplicates —
   `DISTINCT` ensures it's only counted once.)

**Key SQL patterns used:**
| Pattern | When to use |
|---------|------------|
| `GROUP BY` multiple cols | Detecting duplicates, grouping by composite key |
| `HAVING COUNT(*) > 1` | Filter AFTER aggregation (WHERE filters BEFORE) |
| Subquery as table alias | Chain aggregations — first find groups, then count them |
| `COUNT(DISTINCT col)` | Count unique values, not total rows |

**Interview tip:** always clarify "what counts as a duplicate?" before writing the query.
Here it's all 3 fields (company + title + description), but in some problems it might
be just 2 fields.
